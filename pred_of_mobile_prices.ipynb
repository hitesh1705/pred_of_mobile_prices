```python
feature_cols=['battery_power','blue',	'clock_speed',	'dual_sim',	'fc',	'four_g',	'int_memory',	'm_dep',	'mobile_wt',	'n_cores',	'pc',
              'px_height',	'px_width',	'ram',	'sc_h',	'sc_w',	'talk_time',	'three_g',	'touch_screen',	'wifi'	]
x=data.loc[:,data.columns != 'price_range']
y=data['price_range']
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
def plotcnf(cnf_matrix,title):
  class_names=[0,1,2,3]
  fig, ax = plt.subplots()
  tick_marks = np.arange(len(class_names))
  plt.xticks(tick_marks, class_names)
  plt.yticks(tick_marks, class_names)
  sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
  ax.xaxis.set_label_position("top")
  plt.tight_layout()
  plt.title(title, y=1.1)
  plt.ylabel('Actual label')
  plt.xlabel('Predicted label')
```

# Decision Tree Classifier

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=30)

from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier(random_state=1)

dt_model.fit(x_train,y_train)
from sklearn.metrics import accuracy_score
import sklearn.metrics as metrics
y_pred=dt_model.predict(x_test)
print('Accuracy =',(accuracy_score(y_pred,y_test))*100)
plotcnf(cnf_matrix=metrics.confusion_matrix(y_pred,y_test,),title='Confusion Matrix of Decision Tree')
print("Classification report :\n",metrics.classification_report(y_test,y_pred))
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

clf=RandomForestClassifier(n_estimators=100)
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)

print("Accuracy =",(metrics.accuracy_score(y_test, y_pred))*100)
plotcnf(cnf_matrix=metrics.confusion_matrix(y_pred,y_test,),title='Confusion Matrix of Random Forest')
print("Classification report :\n",metrics.classification_report(y_test,y_pred))
```

# Logistic Regression Classifier

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(solver='liblinear')
logreg.fit(x_train,y_train)
y_pred=logreg.predict(x_test)
print('Accuracy =',(accuracy_score(y_pred,y_test))*100)
plotcnf(cnf_matrix=metrics.confusion_matrix(y_pred,y_test,),title='Confusion Matrix of Logistic Regression classfier')
print("Classification report :\n",metrics.classification_report(y_test, y_pred))
```

# KNN Classification

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
n_neighbors=9
knn=KNeighborsClassifier(n_neighbors)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
print('Accuracy =',(accuracy_score(y_pred,y_test))*100)
plotcnf(cnf_matrix=metrics.confusion_matrix(y_pred,y_test,),title='Confusion Matrix of KNN')
print("classification report :\n",metrics.classification_report(y_test, y_pred))
```

# Naive Bayes Method

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
y_pred = gnb.fit(x_train, y_train).predict(x_test)

print('Accuracy =',(accuracy_score(y_pred,y_test))*100)
plotcnf(cnf_matrix=metrics.confusion_matrix(y_pred,y_test,),title='Confusion Matrix of Naive BayesS')
print("classification report :\n",metrics.classification_report(y_test, y_pred))
```
